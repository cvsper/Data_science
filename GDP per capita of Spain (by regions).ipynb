```python
#import Seaborn, Matplotlib, & Pandas
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#initiate data in pandas
data = pd.read_csv('GDP_capita_spanish_regions.csv')
```

```python
#Make data a boxplot
data.plot()
```

```python
#Add x & y labels
plt.xlabel('YEAR')
plt.ylabel('GDP')
plt.title('GDP per capita of Spain (by regions)')
plt.style.use('ggplot')
```

```python
#Save and view your plot
plt.savefig('lineplot.png')
plt.show()
```
