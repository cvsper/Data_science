```python
import seaborn as sns
import pandas as pd
import sklearn as sk
from sklearn import preprocessing, cross_validation, svm
from sklearn.linear_model import LinearRegression as LR
from sklearn.neighbors import KNeighborsClassifier as kn
import numpy  as np
import matplotlib.pyplot as plt 
import quandl as Ql
import math,datetime
import matplotlib.pyplot as plt
from matplotlib import style

plt.style.use('ggplot')
```

```python
data = Ql.get('BITSTAMP/USD') 
data2 = pd.read_csv('homelessness/Population-by-state.csv')
```

```python
# First 5 rows of the data
data.tail()
```

```python
# Grab data intel
intel = ['High', 'Low', 'Volume', 'Last']
data = data[intel]
data['HL_percent'] = (data['High'] - data['Low']) / data['Low'] * 100.0
data['HL_change'] = (data['High'] - data['Low']) / data['High'] * 100.0
intel2 = ['HL_change', 'HL_percent', 'Volume', 'Last']
data = data[intel2]
data.head()
```

```python
forcast_col = 'Last'
forcast_out = int(math.ceil(0.1 * len(data)))
```

```python
data['Label'] = data[forcast_col].shift(-forcast_out)
```

```python
data.fillna(-99999, inplace=True)
```

```python
# 1. Set up variables and Preprocess
daty = data['Label']
datx = data.drop(['Label'],1)
X = datx
y = daty
X = preprocessing.scale(X)
y = np.array(data['Label'])
# 2. Initiate linear regression
clr = LR(n_jobs=-1)
print(len(X),len(y))

#3. make splits and fit
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,y, test_size=0.4)
clr.fit(X_train,y_train)

# 4 Test you results
accuracy = clr.score(X_test, y_test)
forcas_set = clr.predict(X)
```

```python
print(accuracy)
print(forcast_out)

print('Days ahead :', forcast_out, 'Percentage :', accuracy)
```
